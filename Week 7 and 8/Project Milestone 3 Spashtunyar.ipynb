```python
#First step is getting the data into datatables
import requests as r
import pandas as pd
import xlrd
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt
```

```python
url1 = "https://en.wikipedia.org/wiki/List_of_largest_meteorites_on_Earth"
```

```python
list = pd.read_html(url1)
```

```python
Iron = list[0]
Iron
```

```python
StonyIron = list[1]
```

```python
StonyIron
```

```python
#First step is loading my data which consists of two datatables which have different information off the wikipedia site
```

```python
#Action 1 Rename the TKW column to MASS so I can have an easier data join
StonyIron = StonyIron.rename(columns={"TKW": "Mass"})
StonyIron.columns
```

```python
#Action 2 combine the two
frames = [Iron, StonyIron]
HTMLSite = pd.concat(frames)
```

```python
HTMLSite
```

```python
#Action 3, remove the image column, as far as data goes, it serves no use for me and is all NaN anyways
HTMLSite = HTMLSite.dropna(axis=1, how='all')
```

```python
#Verify
HTMLSite.columns
```

```python
#Action 4 parse out Mass so we can get a sort going
HTMLSite['Mass lbs'] = HTMLSite['Mass'].apply(lambda st: st[st.find("(")+1:st.find(")")])
```

```python
#Action 5 removing the units from Mass lbs to get something I can make numerical
HTMLSite['Mass lbs'] = HTMLSite['Mass lbs'].str.rstrip(' lb')
HTMLSite
```

```python
#Action 6 remove the comma
HTMLSite['Mass lbs'] = HTMLSite['Mass lbs'].str.replace(',', '').astype(float)
```

```python
#Action 7 Redo the sorting so that they are by MASS of the two combined tables
HTMLSite.sort_values('Mass lbs', ascending=False)
```

```python
HTMLSite.columns
```

```python
#Action 8 setting new index to N as largest to smallest
HTMLSite.set_index(['N°'])
```

```python
#Action 9, coordinates for one of the values did not load properly so we are loading it back in
HTMLSite.iloc[0,[4]] = "19°35′32″S / 17°56′01″E"
```
