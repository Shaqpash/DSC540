```python
#Creating the top function based on net clout
def top(df, n = 5, column = 'NET CLOUT'):
    return df.sort_values(by=column)[-n:]
```

```python
#Top 6 clout candy bars
top(Candydata, n=6)
```

```python
#Excercise 7 we apply the top function to group values by JOY with clout from lowest to higest
Candydata.groupby('JOY').apply(top)
```
